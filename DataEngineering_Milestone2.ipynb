```python
# Dropping the old SCHEDULED_ARRIVAL & ARRIVAL_TIME columns:
dfr1=dfr1.drop(['SCHEDULED_ARRIVAL','ARRIVAL_TIME'],axis=1)
dfr1
```

```python
# Changing the DIVERTED & CANCELLED into boolean instead of integar
possibility = {1: 'TRUE', 0: 'FALSE'}
dfr1=dfr1.replace({"DIVERTED": possibility})
dfr1=dfr1.replace({"CANCELLED": possibility})
dfr1
```

```python
dfr1['ARRIVAL_HOUR']=dfr1['ARRIVAL_HOUR'].astype(int)
dfr1['ARRIVAL_HOUR']=dfr1['ARRIVAL_HOUR'].astype(str)

dfr1['ARRIVAL_MINUTE']=dfr1['ARRIVAL_MINUTE'].astype(int)
dfr1['ARRIVAL_MINUTE']=dfr1['ARRIVAL_MINUTE'].astype(str)

dfr1['SCHEDULED_ARRIVAL_HOUR']=dfr1['SCHEDULED_ARRIVAL_HOUR'].astype(int)
dfr1['SCHEDULED_ARRIVAL_HOUR']=dfr1['SCHEDULED_ARRIVAL_HOUR'].astype(str)

dfr1['SCHEDULED_ARRIVAL_MINUTE']=dfr1['SCHEDULED_ARRIVAL_MINUTE'].astype(int)
dfr1['SCHEDULED_ARRIVAL_MINUTE']=dfr1['SCHEDULED_ARRIVAL_MINUTE'].astype(str)

dfr1['TimeOFArrival']= dfr1['ARRIVAL_HOUR']+':'+dfr1['ARRIVAL_MINUTE'].apply(str) 
dfr1['TimeOFScheduledArrival']= dfr1['SCHEDULED_ARRIVAL_HOUR']+':'+dfr1['SCHEDULED_ARRIVAL_MINUTE'].apply(str) 

dfr1=dfr1.drop(['ARRIVAL_HOUR','ARRIVAL_MINUTE','SCHEDULED_ARRIVAL_HOUR','SCHEDULED_ARRIVAL_MINUTE'],axis=1)
dfr1
```

```python
dfr1['Early']= (dfr1['ARRIVAL_DELAY']> 0)
dfr1['Late']= (dfr1['ARRIVAL_DELAY']<0)
dfr1
```

```python
dfr1['Early_by']=dfr1['ARRIVAL_DELAY']
dfr1['Late_by']=dfr1['ARRIVAL_DELAY']
dfr1
```

```python
dfr1['Early_by'] = np.where(dfr1['Early_by'] <= 0, 0, dfr1['Early_by'])
dfr1['Late_by'] = np.where(dfr1['Late_by'] >= 0, 0, dfr1['Late_by'])
dfr1['Late_by']=dfr1['Late_by']*-1
dfr1=dfr1.drop(['ARRIVAL_DELAY'],axis=1)
dfr1
```
