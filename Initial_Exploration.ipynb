**Adjusting values in the data**

```python
# Generalizing the price to be in dollars:
# Checking what type of currencies do we have in the dataset -> PLN=0.2$ EUR=1.05$
data.Currency.unique()
#Converting the price column based on the currency 
(data['Price'][data['Currency'] == 'PLN'])= (data['Price'][data['Currency'] == 'PLN'])*0.2  
(data['Price'][data['Currency'] == 'EUR'])= (data['Price'][data['Currency'] == 'EUR'])*1.05                                                                 

data
```

```python
# Changing the currency to dollars
data['Currency'] = data['Currency'].replace('PLN','Dollar')
data['Currency'] = data['Currency'].replace('EUR','Dollar')
display(data)
```
